1. read csv of publication information that contains only scopus IDs

```python
publication_rows = []
with open("publications.csv", "r") as f:
    r = csv.reader(f)
    publication_rows = [row for row in r]
pub_col_names = publication_rows[0]
pub_col_names.append("First Author Member Name")
pub_col_names.append("Last Author Member Name")
pub_col_names.append("Corresponding Author Member Name")
publication_rows = publication_rows[1:]
```

```python
ids = []

with open("output.csv", "r") as f:
    r = csv.reader(f)
    ids = [row[3] for row in r]
ids = ids[1:]

names = []
with open("output.csv", "r") as f:
    r = csv.reader(f)
    names = [f"{row[1]} {row[0]}" for row in r]
names = names[1:]
```

```python
print(type(ids[0]))
```

```python
def append_name_to_row(pub_row:list, memberid:int) -> list:
    try:
        member_name = names[ids.index(memberid)]
    except:
        member_name = "Error!"
    pub_row.append(member_name)
    return pub_row
```

```python
# target cols 11, 12, 13
first_author_rows = []
last_author_rows = []
corresponding_author_rows = []
all_matches = []

for pub_row in publication_rows:
    match_found = False
    first_auth = pub_row[11]
    last_auth = pub_row[12]
    corr_auth = pub_row[13]
    if first_auth in ids:
        match_found = True
        pub_row = append_name_to_row(pub_row, first_auth)
        first_author_rows.append(pub_row)
    else:
        pub_row.append("")
        
    if last_auth in ids:
        match_found = True
        pub_row = append_name_to_row(pub_row, last_auth)
        last_author_rows.append(pub_row)
    else:
        pub_row.append("")
    
    if "|" in corr_auth:
        multi_authors = [au_id.strip() for au_id in corr_auth.split("|")]
        print(multi_authors)
        member_name = ""
        for author in multi_authors:
            if author in ids:
                member_name = names[ids.index(author)] + ", " + member_name
        pub_row.append(member_name)
        corresponding_author_rows.append(pub_row)
    elif corr_auth in ids:
        match_found = True
        pub_row = append_name_to_row(pub_row, corr_auth)
        corresponding_author_rows.append(pub_row)
    else:
        pub_row.append("")

    if match_found == True:
        all_matches.append(pub_row)
```

```python
with open("publications_all_matches.csv", "w+") as f:
    w = csv.writer(f)
    w.writerow(pub_col_names)
    w.writerows(all_matches)
```
